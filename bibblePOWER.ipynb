```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pos_word = pd.read_csv('posWords.csv') 
neg_word = pd.read_csv('negWords.csv')
```

```python
file = open('barbie_fairytopia.txt') #put movie file here
script = file.read()
```

```python
#change the negative word list dataframe into a list, creating an empty list for the negative word occurances    
num_pos = []
num_neg = []
pos_word_list = pos_word['pos_words'].tolist()
neg_word_list = neg_word['neg_words'].tolist()
```

```python
ty = script.split() 

for i in pos_word_list:
    num = ty.count(i)
    num_pos.append(num)
    
for g in neg_word_list:
    num = ty.count(g)
    num_neg.append(num)
```

```python
#add a new column to the negative word list that counts the occurances of the word in script 
pos_word['occurances'] = num_pos
neg_word['occurances'] = num_neg
pd.set_option("display.max_rows", None)
```

```python
#total number of negative and positive words:
tpos = sum(num_pos)
tneg= sum(num_neg)

#total number of words in script:
tw = len(script)

#percentage of negative words:
per_neg = sum(num_neg)/len(script)

#percentage of positive words:
per_pos = sum(num_pos)/len(script)

#difference
d = tpos -tneg

#abosolute diff (p-n)/(total)
pd = (tpos - tneg)/len(script)

#relative diff (p-n)/(p+n)
rd = (tpos - tneg)/(tpos + tneg)

print('positive:',tpos)
print('negative:',tneg)
print('total words:',tw)
print('percentage of negative words:',per_neg)
print('percentage of positive words:',per_pos)
print('difference:', d)
print('absolute diff. score:', pd)
print('relative diff. score:', rd)
```
